# To Procure Telemetry Datatable

```python
season = 2025
telemetry_data = []

for round_no in range(1, 15):  # Races 1 through 14
    print(f"Processing Round {round_no}...")
    
    # Load race session
    session = fastf1.get_session(season, round_no, 'R')
    session.load()
    
    # Loop through all drivers
    for drv in session.drivers:
        laps = session.laps.pick_drivers(drv)
        if laps.empty:
            continue
        
        # Pick fastest lap for this driver
        fastest_lap = laps.pick_fastest()
        if fastest_lap is None:
            continue
        
        tel = fastest_lap.get_car_data().add_distance()
        
        # Add telemetry data row by row
        for _, row in tel.iterrows():
            telemetry_data.append({
                'raceId': f"{season}_R{round_no:02d}",
                'driverId': fastest_lap['Driver'],
                'driverName': fastest_lap['Driver'],
                'lapNumber': fastest_lap['LapNumber'],
                # Convert to seconds with 4 decimals
                'sessionTime': round(row['Time'].total_seconds(), 4),
                'distance': row['Distance'],   # in meters
                'speed': row['Speed'],         # in km/h
                'throttle': row['Throttle'],   # 0–100 %
                'brake': row['Brake'],         # True/False
                'gear': row['nGear'],          # Gear number
                'drs': row['DRS']              # DRS activation (0/1/2)
            })

# Create DataFrame
telemetry_df = pd.DataFrame(telemetry_data)

print("Telemetry data has been successfully downloaded!")
```

# To Procure Tyre Startegies Datatable

```python
season = 2025
lap_tyre_data = []

for rnd in range(1, 15):  # first 14 races
    session = fastf1.get_session(season, rnd, 'R')
    session.load(laps=True, telemetry=False)

    laps = session.laps.copy()

    for idx, row in laps.iterrows():
        lap_tyre_data.append({
            'raceId': f"{season}_R{rnd:02d}",
            'driverId': row['Driver'],
            'lapNumber': row['LapNumber'],
            'compound': row['Compound']
        })

lap_tyre_df = pd.DataFrame(lap_tyre_data)

print(lap_tyre_df.head(20))
```

# To Procure RaceSummary Table

```python
season = 2025
race_summary_data = []

for rnd in range(1, 15):  # Races 1–14
    print(f"Processing Round {rnd}...")

    session = fastf1.get_session(season, rnd, 'R')
    session.load(laps=True, telemetry=False)

    results = session.results
    laps = session.laps

    for _, row in results.iterrows():
        drv = row['Abbreviation']
        fullname = row['FullName'] if 'FullName' in row else row['BroadcastName']
        pos = int(row['Position']) if pd.notnull(row['Position']) else None

        # Pick fastest lap for this driver in the race
        drv_laps = laps.pick_drivers(drv)
        if not drv_laps.empty and 'LapTime' in drv_laps:
            fastest_lap = drv_laps['LapTime'].min()
            if pd.notnull(fastest_lap):
                total_sec = fastest_lap.total_seconds()
                mins = int(total_sec // 60)
                secs = int(total_sec % 60)
                ms = int((total_sec - (mins * 60 + secs)) * 1000)
                fastest_lap_str = f"{mins:02d}:{secs:02d}.{ms:03d}"
            else:
                fastest_lap_str = None
        else:
            fastest_lap_str = None

        race_summary_data.append({
            'raceId': f"{season}_R{rnd:02d}",
            'raceName': session.event['EventName'],
            'driverId': drv,
            'driverName': fullname,
            'finalPosition': pos,
            'fastestLap': fastest_lap_str  # formatted here as mm:ss.mmm
        })

# Build DataFrame
race_summary_df = pd.DataFrame(race_summary_data)

print(race_summary_df.head(10))
```
