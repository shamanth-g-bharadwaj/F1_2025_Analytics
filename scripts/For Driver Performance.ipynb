# To Procure Lap Times Table

```python
lap_times = []

season = 2025
for round_no in range(1, 15):  # Races 1 to 14
    session = fastf1.get_session(season, round_no, 'R')  # Race session
    session.load(laps=True, telemetry=False)  # Load laps but skip telemetry for now
    
    raceId = f"{season}_R{round_no:02d}"
    laps = session.laps
    
    for drv in session.drivers:
        drv_laps = laps.pick_drivers(drv)
        driver_name = session.get_driver(drv)['FullName']
        
        for _, row in drv_laps.iterrows():
            lap_times.append({
                'raceId': raceId,
                'driverId': drv,
                'driverName': driver_name,
                'lapNumber': int(row['LapNumber']),
                'lapTime': str(row['LapTime']) if pd.notnull(row['LapTime']) else None,
                'sector1Time': str(row['Sector1Time']) if pd.notnull(row['Sector1Time']) else None,
                'sector2Time': str(row['Sector2Time']) if pd.notnull(row['Sector2Time']) else None,
                'sector3Time': str(row['Sector3Time']) if pd.notnull(row['Sector3Time']) else None,
                'compound': row['Compound'] if 'Compound' in row and pd.notnull(row['Compound']) else None,
                'stint': int(row['Stint']) if 'Stint' in row and pd.notnull(row['Stint']) else None,
                'pitInLap': bool(row['PitInLap']) if 'PitInLap' in row else False
            })

lap_times_df = pd.DataFrame(lap_times)
print(lap_times_df.head())
```

# To Procure Pit Stops Table

```python
season = 2025
first_round = 1
last_round  = 14  # races completed so far

rows = []

for round_no in range(first_round, last_round + 1):
    session = fastf1.get_session(season, round_no, 'R')
    session.load(laps=True, telemetry=False)

    raceId = f"{season}_R{round_no:02d}"
    laps = session.laps.copy()

    # Safety: ensure sorted by time/lap
    laps = laps.sort_values(['Driver', 'LapNumber']).reset_index(drop=True)

    for drv_num in session.drivers:
        info = session.get_driver(drv_num)             # dict-like
        drv_abbr = info.get('Abbreviation', None)      # e.g., "NOR"
        drv_name = info.get('FullName', info.get('BroadcastName', str(drv_num)))

        if drv_abbr is None:
            # Fallback: try to read from laps
            drv_laps = laps[laps['DriverNumber'] == drv_num]
        else:
            drv_laps = laps[laps['Driver'] == drv_abbr]

        if drv_laps.empty:
            continue

        # Find pit-in laps (where the driver enters the pit at the END of that lap)
        pitin_mask = drv_laps['PitInTime'].notna()
        pitin_idx = drv_laps.index[pitin_mask].tolist()

        stop_no = 1
        for i in pitin_idx:
            in_row = drv_laps.loc[i]

            # Find the first subsequent row for this driver with a PitOutTime
            later = drv_laps.loc[i+1:]  # rows after the pit-in lap
            out_row = later[later['PitOutTime'].notna()].head(1)

            # Defaults
            lap_number = int(in_row['LapNumber']) if pd.notnull(in_row['LapNumber']) else None
            pit_duration_s = None
            compound_after = None
            stint_after = None

            # Duration: next PitOutTime - this PitInTime
            if pd.notnull(in_row['PitInTime']) and not out_row.empty:
                out_time = out_row.iloc[0]['PitOutTime']
                in_time  = in_row['PitInTime']
                # Compute in seconds (guard against weird negatives)
                try:
                    delta = (out_time - in_time).total_seconds()
                    if delta is not None and delta > 0 and delta < 300:  # sanity: < 5 minutes
                        pit_duration_s = round(float(delta), 3)
                except Exception:
                    pass

                # Compound and stint for the OUT-LAP (the lap after the stop)
                compound_after = out_row.iloc[0]['Compound'] if 'Compound' in out_row.columns else None
                if pd.isna(compound_after):
                    compound_after = None

                if 'Stint' in out_row.columns and pd.notnull(out_row.iloc[0]['Stint']):
                    stint_after = int(out_row.iloc[0]['Stint'])
                else:
                    stint_after = None

            rows.append({
                'raceId': raceId,
                'driverId': drv_abbr if drv_abbr is not None else str(drv_num),
                'driverName': drv_name,
                'stop': stop_no,
                'lapNumber': lap_number,
                'pitDurationSeconds': pit_duration_s,   # clean float seconds
                'compound': compound_after,             # compound AFTER stop (out-lap)
                'stint': stint_after                    # stint number after stop
            })
            stop_no += 1

# Build dataframe & save
pit_stops_df = pd.DataFrame(rows)

# Optional: sort for readability
pit_stops_df = pit_stops_df.sort_values(['raceId', 'driverId', 'stop']).reset_index(drop=True)

print(pit_stops_df.head(10))
```

```python
pit_stops_df.to_csv(r"D:\Masters Study Abroad\Self Study\Personal Projects\F1\Descriptive Analytics\PitStops_2025.csv", index=False)
```

# To Procure Driver Standings Table

```python
import pandas as pd

# Load the Race Results table you built earlier
results_df = pd.read_csv(r"D:\Masters Study Abroad\Self Study\Personal Projects\F1\Descriptive Analytics\RaceResults_2025.csv")

standings_list = []

# Process round by round
for rnd in sorted(results_df['raceId'].unique()):
    # Subset results up to and including this race
    up_to_round = results_df[results_df['raceId'] <= rnd]
    
    # Compute cumulative points per driver
    cum_points = (up_to_round.groupby(['driverId', 'driverName', 'teamName'])['points']
                  .sum()
                  .reset_index())
    
    # Add race/round info
    cum_points['raceId'] = rnd
    cum_points['season'] = 2025
    cum_points['round'] = int(rnd.split('_R')[-1])
    
    # Rank by points to get championship positions
    cum_points['championshipPosition'] = cum_points['points'].rank(
        method='first', ascending=False
    ).astype(int)
    
    # Rename for clarity
    cum_points.rename(columns={'points': 'pointsAfterRace'}, inplace=True)
    
    standings_list.append(cum_points)

standings_df = pd.concat(standings_list).sort_values(['round', 'championshipPosition'])
print(standings_df.head(20))
```

```python
standings_df.to_csv(r"D:\Masters Study Abroad\Self Study\Personal Projects\F1\Descriptive Analytics\DriverStandings_2025.csv", index=False)
```

# To Procure Lap by Lap Positions Table

```python
season = 2025
first_round, last_round = 1, 14

rows = []

for round_no in range(first_round, last_round + 1):
    session = fastf1.get_session(season, round_no, 'R')
    session.load(laps=True, telemetry=False)

    raceId = f"{season}_R{round_no:02d}"
    laps = session.laps.copy()

    # Keep only the columns we need; guard for missing ones
    keep_cols = ['Driver', 'DriverNumber', 'LapNumber', 'Position',
                 'PitInTime', 'TrackStatus']
    for col in keep_cols:
        if col not in laps.columns:
            laps[col] = pd.NA

    # Ensure sorted correctly for deltas
    laps = laps.sort_values(['Driver', 'LapNumber']).reset_index(drop=True)

    # Map driver number -> meta
    # (session.get_driver works with driver number)
    drvnum_to_meta = {}
    for drv_num in session.drivers:
        meta = session.get_driver(drv_num)
        drvnum_to_meta[drv_num] = {
            'abbr': meta.get('Abbreviation', None),
            'name': meta.get('FullName', meta.get('BroadcastName', str(drv_num)))
        }

    # We’ll compute delta per driver
    for drv_num in session.drivers:
        meta = drvnum_to_meta[drv_num]
        abbr = meta['abbr']

        # Use the abbr if present; fall back to number match
        if abbr:
            dlaps = laps[laps['Driver'] == abbr].copy()
        else:
            dlaps = laps[laps['DriverNumber'] == drv_num].copy()

        if dlaps.empty:
            continue

        # Clean numeric position
        # (Some rows may be NaN during pits/SC; keep them but handle in delta)
        dlaps['position'] = pd.to_numeric(dlaps['Position'], errors='coerce')

        # Delta: previous_position - current_position
        dlaps['prev_pos'] = dlaps['position'].shift(1)
        dlaps['deltaPosition'] = dlaps['prev_pos'] - dlaps['position']
        # First lap has no previous; set delta to 0 for convenience
        dlaps.loc[dlaps['prev_pos'].isna(), 'deltaPosition'] = 0
        dlaps['deltaPosition'] = dlaps['deltaPosition'].astype('float').round(0)

        # Pit flag: this lap ended with a pit entry if PitInTime is not null
        dlaps['isPitLap'] = dlaps['PitInTime'].notna()

        # Build rows
        for _, r in dlaps.iterrows():
            rows.append({
                'raceId': raceId,
                'driverId': abbr if abbr else str(drv_num),
                'driverName': meta['name'],
                'lapNumber': int(r['LapNumber']) if pd.notnull(r['LapNumber']) else None,
                'position': int(r['position']) if pd.notnull(r['position']) else None,
                'deltaPosition': int(r['deltaPosition']) if pd.notnull(r['deltaPosition']) else 0,
                'isPitLap': bool(r['isPitLap']),
                'trackStatus': r['TrackStatus'] if pd.notnull(r['TrackStatus']) else None
            })

# Assemble & save
positions_df = pd.DataFrame(rows).sort_values(
    ['raceId', 'driverId', 'lapNumber']
).reset_index(drop=True)

print(positions_df.head(12))
```
